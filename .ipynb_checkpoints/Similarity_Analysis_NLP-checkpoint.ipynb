```python
import pandas as pd
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000) 
%matplotlib inline
```

### Step 1: Build Corpus of Documents. Created 7 documents and labels as shown in the code below:

```python
corpus = ['Parasite movie won oscar awards','Funny, touching movie, Parasite movie, Oscar winner 2020',
          'Tesla electric cars are the future','Never skip breakfast, not good for health',
         'Breakfast is the important meal of the day, never skip','Tesla electric cars beat any car in the market',
         'Good movies win oscars, Parasite Oscar movie', 'Cannot compete with Tesla, they are the best electric cars']
```

```python
labels = ['movie','movie','cars','breakfast','breakfast','cars','movie','cars']
```

Create a dafaframe (table format) with Documents and Labels

```python
corpus = np.array(corpus)
corpus_df = pd.DataFrame({'Document':corpus,'Category':labels})
corpus_df = corpus_df[['Document','Category']]
corpus_df
```

```python
wpt = nltk.WordPunctTokenizer()
stop_words = nltk.corpus.stopwords.words('english')
```

Function to normalize the documents. The below function will:
* remove special characters
* extra white spaces, digits
* stopwords
* lower casing the entire corpus

```python
def normmalize_document(doc):
    doc = re.sub(r'[^a-zA-Z\s]', '', doc, re.I|re.A )
    doc = doc.lower()
    doc = doc.strip()
    tokens = wpt.tokenize(doc)
    filtered_tokens = [token for token in tokens if token not in stop_words]
    doc = ' '.join(filtered_tokens)
    return doc

normalize_corpus = np.vectorize(normmalize_document)
```

```python
norm_corpus = normalize_corpus(corpus)
norm_corpus
```

Removed, stop words, special characters, lowered cases and stripped extra white spaces!

### Step 2: Perform Term-Frequency with a table output:

Note: Normalizing is already done above. Not required to perform again in this step. 

```python
from sklearn.feature_extraction.text import CountVectorizer
# get bag of words features in sparse format
cv = CountVectorizer(min_df=0., max_df=1.)
cv_matrix = cv.fit_transform(norm_corpus)
cv_matrix
```

Transforming Sparse Matrix to Dense Matrix, so the output is more readable as shown below:

```python
cv_matrix = cv_matrix.toarray()
cv_matrix
```

Creating the table output (DataFrame) to view the Document-Frequencies:

```python
#get all unique words in the corpus
vocab = cv.get_feature_names()
#show document feature vectors
pd.DataFrame(cv_matrix, columns=vocab)
```

From the table above it can be noted that, the words appear in the document are counted as per their occurence. For example, in the first document, words such as "awards", "movie", "oscar" are mentioned once and hence represented as 1 in the dataframe. The same process is followed for all the documents. The above table is also called as "Document Frequency"

Let's compute the <b>Term Frequency</b> for the corpus now:
* Step 1: Count the number of unique words in the corpus
* Step 2: Calculate the Sum of the Document Frequency (column wise) --> sum_terms
* Step 3: Get the number of unique words --> No_of_Words
* Step 4: Calculate Term Frequencies, by calculating  -->  sum_terms/No_of_Words

```python
#get all unique words in the corpus
vocab = cv.get_feature_names()
#show document feature vectors
term_freq = pd.DataFrame(cv_matrix, columns=vocab)
sum_terms = pd.DataFrame(np.sum(term_freq))
sum_terms = sum_terms.T
sum_terms
```

```python
No_of_Words = len(sum_terms.columns)
No_of_Words
```

```python
Term_Frequency = sum_terms/No_of_Words
Term_Frequency
```

Term Frequencies are calculated as per the steps mentioned above. It can be noted that, terms "movie", "cars", 
"electric", "parasite","tesla" appears more frequently in the corpus than any other terms and hence have an higher Term Frequency Score than other words in the corpus.

### Step 3. Perform TF-IDF (normalized corpus) and create a table with scores and vocabulary items 

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tt = TfidfTransformer(norm='l2',use_idf = True)
tt_matrix = tt.fit_transform(cv_matrix)

tt_matrix = tt_matrix.toarray()
vocab = cv.get_feature_names()
pd.DataFrame(np.round(tt_matrix,2), columns = vocab)
```

The above table contains TF-IDF scores for each vocabulary items. It provides more weightage to certain words compared to other words in the same document. For example, in the first row of the dataframe above, "awards" column has a TF-IDF score of 0.62 compared to the term "movie" or "oscar" or "parasite". Even though the words movie, oscar and parasite occur more frequently, they receive a lower TF-IDF score compared to the word "awards". This is because the word "awards" is unique in the corpus, TF-IDF assigns a higher score to unique words in comparison to common words by which we are retaining or obtaining more information from the dataset compared to just term-frequencies or bag of words model.

### Step 4. Perform Cosine Similarity metrics between 3 documents


```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(tt_matrix)
similarity_df = pd.DataFrame(similarity_matrix)
similarity_df
```

The above dataframe prints the Cosine Similarity Metrics. It can be noted that, (0,1) and (0,6) for the first row has high scores compared to other documents (Ignore diagonal scores such as (0,0),(1,1) etc.. as they are the same document and their similarity score will always be 1). From the labels, we know this is True, as these are related to the same document --> 'movie'. Similar charecteristics can be inferred for other rows as well. 

### Step 5. Perform Document clustering, create a table with distance and cluster size, draw a dendrogram

```python
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(similarity_matrix, 'ward')
pd.DataFrame(Z,columns=['Document\Cluster 1', 'Document\Cluster 2', 'Distance','Cluster Size'],dtype='object')
```

Create Dendrogram plot:

```python
plt.figure(figsize=(8,3))
plt.title('Hierarchical Clustering Dendogram')
plt.xlabel('Data point')
plt.ylabel('Distance')
dendrogram(Z)
plt.axhline(y=1.0,c='k',ls='--',lw=0.5)
```

From the dendrogram above, we can notice that each data points are being considered as individual clusters. Then they slowly merge based on the similarity scores. For example, (0,1) clusters has the lower distance and hence they appear to be more closer in contents and hence they are merged together. Similarly, (6,(0,1) are merged together as they appear to be the next closest in the cluster. The same approach is followed to all the datapoints based on the similarity score. A line at Distance = 1, clearly shows that there are "three" identical clusters in the document which we will label in the next section.

### Step 6. Create a table with labels and clusters labels.

```python
from scipy.cluster.hierarchy import fcluster
max_dist = 1.0

cluster_labels = fcluster(Z,max_dist,criterion='distance')
cluster_labels = pd.DataFrame(cluster_labels,columns=['ClusterLabel'])
pd.concat([corpus_df,cluster_labels],axis = 1)
```

### Explanation of Results:

We pre-processed the corpus by removing stop words, unwanted white spaces and special characters. We calcualted Document Frequencies, Term Frequencies and TF-IDF Scores. Also, we performed similarity analysis between the documents and classified into a hierarchial cluster as mentioned in the lecture. The above dataframe contains three columns. they are:

* "Document"     --> Content in the document
* "Category"     --> Labels of the document
* "ClusterLabel" --> Cluster Classification

From the above dataframe, we shall infer that the cluster analysis have successfully identified "three" clusters and is clearly indicated by the ClusterLabel column. For example, the category "movie" is accurately classifed as "3" in the "ClusterLabel"  column. Similarly for "cars" --> 1 and "breakfast" --> 2 are accurately classified as well

In summary, the algorithm has succesfully identified <b>three distinct categories in our corpus</b> and accurately matches with the labels we created in the beginning of the section.
